````python
import pandas as pd
import yaml
import ast
import sys
import os

from outfitgenerator import OutfitGenerator
from resources.utils import plot_array_img

with open('/content/drive/MyDrive/boost_zara/resources/config_boost2025.yml', 'r') as file: config = yaml.safe_load(file)
with open('/content/drive/MyDrive/boost_zara/resources/heros_challenge.yml', 'r') as file: heros_challenge = yaml.safe_load(file)

za_kid_catalog = pd.read_csv('/content/drive/MyDrive/boost_zara/resources/catalog_boost2025.csv', index_col=0)
za_kid_catalog["LICENSE_TAG"] = (za_kid_catalog["LICENSE_TAG"].str.split(",").fillna("").map(set))
za_kid_catalog["COMERCIAL_RULES"] = za_kid_catalog["COMERCIAL_RULES"].fillna("")

outfits = pd.read_csv('/content/drive/MyDrive/boost_zara/resources/outfits_boost2025.csv', index_col=0)
outfits['OUTFIT_PARTNUMBER_LIST'] = outfits['OUTFIT_PARTNUMBER_LIST'].apply(ast.literal_eval)
outfits['OUTFIT_PHOTO_PLAIN_LIST'] = outfits['OUTFIT_PHOTO_PLAIN_LIST'].apply(ast.literal_eval)

og = OutfitGenerator(config=config)
````
